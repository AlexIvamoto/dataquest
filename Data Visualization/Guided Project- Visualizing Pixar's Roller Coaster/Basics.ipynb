```python
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str[:-2].astype('float')
pixar_movies['Domestic %']
```

```python
pixar_movies['International %'] = pixar_movies['International %'].str[:-2].astype('float')
pixar_movies['International %']
```

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score'] * 10
pixar_movies['IMDB Score']
```

```python
filtered_pixar = pixar_movies[0:-1]
filtered_pixar
```

```python
pixar_movies.set_index(['Movie'], inplace=True)
pixar_movies
```

```python
filtered_pixar.set_index(['Movie'], inplace=True)
filtered_pixar
```

```python
critics_reviews = pixar_movies[['RT Score', 'IMDB Score', 'Metacritic Score']]
critics_reviews
```

```python
critics_reviews.plot()
sns.plt.show()
```

```python
critics_reviews.plot(figsize=(6,10))
sns.plt.show()
```

```python
critics_reviews.plot(kind='box', figsize=(9,5))
sns.plt.show()
```

```python
revenue_proportions = filtered_pixar[['Domestic %', 'International %']]
revenue_proportions
```

```python
revenue_proportions.plot(kind='bar', stacked=True)
sns.plt.show()
```

```python
revenue_proportions.plot(kind='area')
sns.plt.show()
```

```python
oscar = filtered_pixar[['Oscars Nominated', 'Oscars Won']]
oscar
```

```python
oscar.plot(kind='bar')
sns.plt.show()
```
